```python
f = open('sample.txt', mode = 'r')
res = f.readlines()
f.close()
res = res[1:-1]
for i in range(len(res)):
    res[i] = float(res[i])
print(res)
```

```python
def avg_list(I):
    sum = 0
    for current in I:
        sum += current
    return sum / len(I)
```

```python
avg_list(res)
```

```python
f = open('res.txt', mode = 'w')
f.write('avg = '+str(avg_list(res)))
f.close()
```
