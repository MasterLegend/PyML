```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 5)
y = 2 * x + 1
```

```python
x, y
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y, s = 200)
plt.show()
```

```python
plt.scatter(x, y, s = 200, marker = 's')
plt.show()
```

```python
plt.scatter(x, y, s = 200, marker = 's', c = 'red')
plt.show()
```

```python
plt.xlim(0, 30)
plt.ylim(0, 50)

plt.scatter(x, y, s = 200, marker = 's', c = 'red')
plt.show()
```

```python
plt.xlim(0, 30)
plt.ylim(0, 50)

x_ticks = np.linspace(0, 30, 11)
plt.xticks(x_ticks)

plt.scatter(x, y, s = 200, marker = 's', c = 'red')
plt.show()
```

```python
plt.xlim(0, 30)
plt.ylim(0, 50)

x_ticks = np.linspace(0, 30, 11)
plt.xticks(x_ticks)

plt.xlabel('x_label')
plt.ylabel('y_label')

plt.scatter(x, y, s = 200, marker = 's', c = 'red')
plt.show()
```

```python
x = np.linspace(-10, 10, 40)
y = x * x + 1
plt.plot(x, y, c='red')
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
y = np.sin(x)

plt.xlim(-5, 5)
plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3]
y = [5, 3, 4]
plt.bar(x, y, width = 0.2, color = 'r')
plt.show()
```
