```python
from matminer.datasets import load_dataset

df = load_dataset('castelli_perovskites',data_home = '.')
df
```

```python
df = df[['formula', 'structure', 'e_form']]
df
```

```python
from matminer.featurizers.structure.matrix import CoulombMatrix
from matminer.featurizers.structure.matrix import SineCoulombMatrix
```

```python
cm = CoulombMatrix()
```

```python
df1 = cm.fit_featurize_dataframe(df, 'structure')
```

```python
df1
```

```python
scm = SineCoulombMatrix()
df2 = scm.fit_featurize_dataframe(df, 'structure')
```

```python
df2
```

```python
X = df2.drop(['formula', 'structure', 'e_form'], axis = 1).values
y = df2['e_form'].values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ssx = StandardScaler()
X_s = ssx.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_s, y, random_state = 1)
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(objective = 'reg:squarederror')
xgb.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (5, 5))
plt.scatter(y_train, xgb.predict(X_train),c = 'b', s = 1, alpha =0.5)
plt.scatter(y_test, xgb.predict(X_test),c = 'r', s = 1, alpha =0.5)
plt.show()
```

```python
xgb.score(X_train, y_train), xgb.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 3)
X_p = pca.fit_transform(X)
```

```python
X_p
```

```python
ssx2 = StandardScaler()
X_ps = ssx2.fit_transform(X_p)
```

```python
X_ps
```

```python
X_trainp, X_testp, y_trainp, y_testp = train_test_split(X_ps, y, random_state = 1)
```

```python
xgb2 = XGBRegressor(objective = 'reg:squarederror')
xgb2.fit(X_trainp, y_trainp)
```

```python
plt.figure(figsize = (5, 5))
plt.scatter(y_trainp, xgb2.predict(X_trainp),c = 'b', s = 1, alpha =0.5)
plt.scatter(y_testp, xgb2.predict(X_testp),c = 'r', s = 1, alpha =0.5)
plt.show()
```

```python
xgb2.score(X_trainp, y_trainp), xgb2.score(X_testp, y_testp)
```
