```python
import pandas as pd

df = pd.read_csv('mp.csv')
df
```

```python
X = df.drop(['formula', 'y'], axis = 1)
y = df['y'].values
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X, y)
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse

print('MAE = %.3f' % (mae(y_true = y, y_pred = lr.predict(X))))
print('MSE = %.3f' % (mse(y_true = y, y_pred = lr.predict(X))))
```

```python
lr.score(X, y)
```

```python
X.shape
```

```python
X_train, y_train = X[:-10], y[:-10]
X_test, y_test = X[-10:], y[-10:]
```

```python
X_train.shape, X_test.shape
```

```python
lr_lo = LinearRegression()
lr_lo.fit(X_train, y_train)
```

```python
import numpy as np
```

```python
R2_lo = lr_lo.score(X_test, y_test)
print(R2_lo)
print('RMSE = %.3f' % np.sqrt(mse(y_true = y_test,
                                 y_pred = lr_lo.predict(X_test))))
```

```python
print(mse(y_true = y_test, y_pred = lr_lo.predict(X_test)))
```

```python
from sklearn.utils import shuffle

epochs = 10
res_rmse = []
for i in range(epochs):
    X_r, y_r = shuffle(X, y)
    X_train, y_train = X_r[:-10], y_r[:-10]
    X_test, y_test = X_r[-10:], y_r[-10:]
    lr_lo = LinearRegression()
    lr_lo.fit(X_train, y_train)
    res_rmse.append(np.sqrt(mse(y_true = y_test, y_pred = lr_lo.predict(X_test))))
    
print(res_rmse)
print(np.mean(res_rmse))
```

```python
from sklearn.model_selection import cross_val_score

rmse_scores = cross_val_score(lr, X, y,
                              scoring = 'neg_root_mean_squared_error', cv = 10)
print('rmse = %.3f' % -np.mean(rmse_scores))
```

```python
rmse_scores
```

```python
from sklearn.model_selection import LeaveOneOut

sample_x = [1, 2, 3, 4, 5]
loo = LeaveOneOut()
for train, test in loo.split(sample_x):
    print(train, test)
```

```python
rmse_scores = cross_val_score(lr, X, y,
                              scoring = 'neg_root_mean_squared_error', cv = len(X))
print('rmse = %.3f' % -np.mean(rmse_scores))
```

```python
from sklearn.utils import resample

X_train, y_train = resample(X, y)
index_test = set(X.index) - set(X_train.index)
X_test = X.iloc[list(index_test)]
y_test = y[list(index_test)]
X_train.shape, X_test.shape
```

```python
lr_bs = LinearRegression()
lr_bs.fit(X_train, y_train)
print('MSE = %.3f' % np.sqrt(mse(y_true = y_test, y_pred = lr_bs.predict(X_test))))
```
